```python
hd_merge = pd.concat([hd_2001,hd_2002,hd_2003,hd_2004,hd_2005,hd_2006,hd_2007,hd_2008,hd_2009,hd_2010,hd_2011,hd_2012,hd_2013,hd_2014,hd_2015,hd_2016,hd_2017,hd_2018,hd_2019,hd_2020,hd_2021],ignore_index=False)
hd_merge = hd_merge.astype(object)

obereg_dict = {0:"US Service schools",1:"New England",2:"Mid East",3:"Great Lakes",4:"Plains",5:"Southeast",6:"Southwest",7:"Rocky Mountains",8:"Far West",9:"Other US jurisdictions"}
sector_dict = {0:"Administrative Unit",1:"4-year public",2:"4-year private, not-for-profit",3:"4-year private, for-profit",4:"2-year public",5:"2-year private, not-for-profit",6:"2-year private, for-profit",7:"Less than 2-year public",8:"Less than 2-year private, not-for-profit",9:"Less than 2-year private, for-profit",99:"sector not known"}
iclevel_dict = {-3:"not available",1:"Four or more years",2:"At least 2 but less than 4 years",3:"Less than 2 years (below associate)"} 
control_dict = { -3:"not available",1:"Public",2:"Private not-for-profit",3:"Private for-profit"}
deggrant_dict = {-3:"not available",1:"Degree-granting",2:"Nondegree-granting, primarily postsecondary",3:"Not primarily postsecondary institutions",4:"Institution is not an educational entity"}
carnegie_dict = {-3:"not available",-2:"not applicable",15:"Doctoral or Research Universities-Extensive",16:"Doctoral or Research Universities-Intensive",21:"Masters Colleges and Universities I",22:"Masters Colleges and Universities II",31:"Baccalaureate Colleges-Liberal Arts",32:"Baccalaureate Colleges-General",33:"Baccalaureate/Associates Colleges",40:"Associates Colleges",51:"Specialized Institutions: Theological seminaries and other special",52:"Specialized Institutions: Medical schools and medical centers",53:"Specialized Institutions: Other separate health profession schools",54:"Specialized Institutions: Schools of engineering and technology",55:"Specialized Institutions: Schools of business and management",56:"Specialized Institutions: Schools of art, music, and design",57:"Specialized Institutions: Schools of law",58:"Specialized Institutions: Teachers colleges",59:"Specialized Institutions: Other specialized institutions",60:"Specialized Institutions: Tribal colleges"}
pset4flg_dict = {1:"Title IV postsecondary institution",2:"Non-Title IV postsecondary institution",3:"Title IV NOT primarily postsecondary institution",4:"Non-Title IV NOT primarily postsecondary institution",5:"Title IV postsecondary institution that is NOT open to the public",6:"Non-Title IV postsecondary institution that is NOT open to the public",9:"Institution is not active in current universe"}
medical_dict = {-2:"not applicable",-1:"not reported",1:"Yes",2:"No"}
hospital_dict = {-3:"not available",-2:"not applicable",-1:"not reported",0:"No",1:"Yes",2:"No"}
openpubl_dict = {-3:"not available",0:"Institution is not open to the public",1:"Insititution is open to the public",2:"Insititution is not open to the public"}

hd_merge = hd_merge.replace({'obereg':obereg_dict,'sector':sector_dict,'iclevel':iclevel_dict,'control':control_dict,'deggrant':deggrant_dict,'carnegie':carnegie_dict,'pset4flg':pset4flg_dict,'medical':medical_dict,'hospital':hospital_dict,'openpubl':openpubl_dict})
hd_merge.head()
```

Add a column for Carnegie_Grouped

```python

def carnegie_grouped(row):
    if row["carnegie"] == 'Doctoral or Research Universities-Intensive' or row["carnegie"] == 'Doctoral or Research Universities-Extensive':
        return 'Doctoral or Research Universities'
    elif row["carnegie"] == 'Masters Colleges and Universities I' or row["carnegie"] == 'Masters Colleges and Universities II':
        return 'Masters Colleges and Universities'
    elif row["carnegie"] == 'Baccalaureate Colleges-General' or row["carnegie"] == 'Baccalaureate Colleges-Liberal Arts':
        return 'Baccalaureate Colleges and Universities'
    elif row["carnegie"] == 'Associates Colleges' or row["carnegie"] == 'Baccalaureate/Associates Colleges':
        return 'Associates Colleges'   
    elif row["carnegie"] == 'Specialized Institutions: Other specialized institutions' or row["carnegie"] == 'Specialized Institutions: Schools of art, music, and design' or row["carnegie"] == 'Specialized Institutions: Schools of business and management' or row["carnegie"] == 'Specialized Institutions: Schools of engineering and technology' or row["carnegie"] == 'Specialized Institutions: Tribal colleges' or row["carnegie"] == 'Specialized Institutions: Medical schools and medical centers' or row["carnegie"] == 'Specialized Institutions: Other separate health profession schools' or row["carnegie"] == 'Specialized Institutions: Schools of law' or row["carnegie"] == 'Specialized Institutions: Teachers colleges' or row["carnegie"] =='Specialized Institutions: Theological seminaries and other special':
        return 'Specialized Institutions'   
    else:
        return 'not applicable'
    
hd_merge['carnegie_grouped'] = hd_merge.apply(lambda row: carnegie_grouped(row), axis=1)

hd_merge.head()
```

Step 4: 
Filter data by scope of the project:
1. Degree-granting
2. Public or not-for-profit
3. Title IV postsecondary institution (accredited)
4. open to the public 

```python
sector_list = ['2-year public','4-year public','4-year private, not-for-profit','Less than 2-year private, for-profit','4-year private, for-profit' '2-year private, not-for-profit','2-year private, for-profit','Less than 2-year public','Less than 2-year private, not-for-profit']
control_list = ['Public','Private not-for-profit']
obereg_list = ['Southeast','Far West','Southwest','Rocky Mountains','Plains','New England','Mid East','Great Lakes']
deggrant_list = ['Degree-granting']
carnegie_grouped_list = ['Associates Colleges','Doctoral or Research Universities','Specialized Institutions','Masters Colleges and Universities','Baccalaureate Colleges and Universities']
pset4flg_list = ['Title IV postsecondary institution']
openpubl_list = ['Insititution is open to the public']

hd_merge_filtered = hd_merge[(hd_merge['sector'].isin(sector_list)) & (hd_merge['control'].isin(control_list)) & (hd_merge['obereg'].isin(obereg_list)) & (hd_merge['deggrant'].isin(deggrant_list)) & (hd_merge['carnegie_grouped'].isin(carnegie_grouped_list)) & (hd_merge['openpubl'].isin(openpubl_list)) & (hd_merge['pset4flg'].isin(pset4flg_list))]

hd_merge_filtered.head()
```

```python
instnm_dict = {}

for index, row in hd_merge_filtered.iterrows():
    unitid = row['unitid']
    instnm = row['instnm']
    year = row['year']
    if row['unitid'] not in instnm_dict.keys():
        instnm_dict[unitid] = {'year': year, 'instnm': instnm.title()}
    elif instnm_dict[unitid]['year'] < year:
        instnm_dict[unitid] = {'year': year, 'instnm': instnm.title()}

id_2_inst_name = {}
for unitid, instnm in instnm_dict.items():
    id_2_inst_name[unitid] = instnm['instnm']

hd_merge_filtered = hd_merge_filtered.copy()

hd_merge_filtered.loc[:,'instnm_normalized'] = hd_merge_filtered['unitid'].map(id_2_inst_name)
hd_merge_filtered.head(2)
```
