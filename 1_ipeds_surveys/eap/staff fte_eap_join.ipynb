```python
import pandas as pd
import datetime as dt
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
eap_2002 = pd.read_csv('eap2002.csv',low_memory=False)
eap_2002 = eap_2002.fillna(0)

eap_2002.info()

functcd = [10,15,16,17,22]
eap_2002 = eap_2002[(eap_2002['fstat']==0) & (eap_2002['functcd'].isin(functcd))]

eap_2002_grouped = eap_2002.groupby(['unitid','functcd','ftpt']).agg({'eaptyp':['max']})
eap_2002_grouped.columns = ['total']
eap_2002_grouped = eap_2002_grouped.reset_index()
```

```python
n = len(pd.unique(eap_2002_grouped['unitid']))
print(n)
```

```python
double = eap_2002_grouped.groupby(['functcd','ftpt']).agg({'unitid':['nunique']})
print(double)
```

```python
eap_2002_grouped.head()
```

```python
eap_2002_pivot = pd.pivot_table(eap_2002_grouped, index=['unitid'],columns=['functcd','ftpt'],values=['total'])
eap_2002_pivot.columns = ['_'.join(str(s).strip() for s in col if s) for col in eap_2002_pivot.columns]
eap_2002_pivot =eap_2002_pivot.reset_index()
eap_2002_pivot['year'] = eap_2002_pivot['year'] = '2002'
eap_2002_pivot = eap_2002_pivot.fillna(0)
eap_2002_pivot.head()
```

```python
merge = pd.concat([eap_2002_pivot,eap_2003_pivot,eap_2004_pivot,eap_2005_pivot,eap_2006_pivot,eap_2007_pivot,eap_2008_pivot,eap_2009_pivot,eap_2010_pivot,eap_2011_pivot],ignore_index=False)
merge = merge[['unitid','year','total_10_1','total_10_2','total_10_3','total_22_1','total_22_2','total_22_3']]
merge = merge.fillna(0)

merge['total non instruction'] = merge['total_10_1'] - merge['total_22_1']
merge['total non instruction ft'] = merge['total_10_2'] - merge['total_22_2']
merge['total non instruction pt'] = merge['total_10_3'] - merge['total_22_3']

merge['total non instruction staff fte'] = merge['total non instruction ft'] + (merge['total non instruction pt']*.333).round(0)
merge['total staff fte'] = merge['total_10_2'] + (merge['total_10_3']*.333).round(0)
merge['total instruction staff fte'] = merge['total_22_2'] + (merge['total_22_3']*.333).round(0)

merge = merge.rename(columns={'total_10_1':'total staff','total_10_2':'total staff ft','total_10_3':'total staff pt','total_22_1':'total instruction staff','total_22_2':'total instruction ft','total_22_3':'total instruction pt'})

merge = merge[['unitid','year','total staff','total staff ft','total staff pt','total staff fte','total instruction staff','total instruction ft','total instruction pt','total instruction staff fte','total non instruction','total non instruction ft','total non instruction pt','total non instruction staff fte']]

merge.info()
```
